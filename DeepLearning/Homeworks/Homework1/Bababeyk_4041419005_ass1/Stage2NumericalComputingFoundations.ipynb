## NumPy Essentials

### What is NumPy?
**NumPy (Numerical Python)** is the core Python library for numerical computing.  
It provides the powerful **`ndarray`** (n-dimensional array) object that lets us store and manipulate numbers efficiently—much faster than plain Python lists.

To use it, import the library (standard alias is `np`):


```python
import numpy as np
```

**Creating Arrays**

A NumPy array is a central data structure in the NumPy library for Python, designed for efficient numerical computations. It represents a multidimensional, homogeneous array of fixed-size items.

Key Characteristics

Structure

- Multidimensional: Can be 1D, 2D, 3D, or higher dimensions

- Homogeneous: All elements must be of the same data type

- Fixed-size: Size is determined at creation time

Performance Advantages

- Fast operations: Implemented in C and optimized for speed

- Vectorized operations: Apply operations to entire arrays without loops

- Memory efficient: Compact storage compared to Python lists

```python
# From a Python list
a = np.array([1, 2, 3, 4])
print("Array a:", a)

# 2-D array (matrix)
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print("\nMatrix b:\n", b)

# Arrays of zeros, ones, or a range
zeros = np.zeros((2,3))
ones  = np.ones((3,3))
arange_example = np.arange(0, 10, 2)   # start, stop, step

print("\nZeros:\n", zeros)
print("\nOnes:\n", ones)
print("\nArange:", arange_example)
```

**Array Shape and Indexing**

```python
print("Shape of b:", b.shape)  # (rows, columns)
print("Element [0, 1]:", b[0, 1])  # 1st row, 2nd col

# Slicing
print("First row:", b[0])
print("First column:", b[:, 0])
```

`.shape` returns a tuple showing dimensions of the array.

**Broadcasting and Vectorized Operations**

Vectorized operations refer to performing operations on entire arrays without using explicit loops. Instead of iterating through each element, operations are applied to the whole array simultaneously.
Key Benefits

- Performance: Much faster than Python loops

- Readability: More concise and readable code

- Maintenance: Easier to write and maintain

```python
x = np.array([1, 2, 3])
y = np.array([10, 20, 30])

print("x + y =", x + y)
print("x * y =", x * y)

# Broadcasting allows operations between arrays of different shapes
m = np.ones((3,3))
v = np.array([1, 2, 3])
print("\nBroadcasted sum:\n", m + v)
```

NumPy automatically broadcasts smaller arrays across larger ones when shapes are compatible.

**Common Array Functions**

```python
data = np.array([[3, 5, 1],
                 [2, 8, 7],
                 [4, 6, 9]])

print("Sum:", data.sum())
print("Mean:", data.mean())
print("Max:", data.max())
print("Column-wise mean:", data.mean(axis=0))
print("Row-wise mean:", data.mean(axis=1))
```

**Random Numbers**

Random numbers are crucial in programming for simulating uncertainty, generating test data, and implementing various algorithms. Python provides robust random number generation capabilities through modules like random and numpy.random.

```python
random_values = np.random.rand(2, 3)   # uniform [0, 1)
random_ints   = np.random.randint(0, 10, size=(2, 3))
print("Random floats:\n", random_values)
print("\nRandom ints:\n", random_ints)
```

Use `np.random` for experiments, simulations, and initializing weights later in PyTorch.

**Practice 1 — NumPy Basics**

Try solving these in a new code cell

1. Create a 3×3 array with values 1 → 9.

2. Compute its row-wise and column-wise means.

3. Generate 100 random integers between 1 and 6 (simulate dice rolls) and count how many times each number appears.

4. Normalize an array so all its values lie between 0 and 1 (hint: (x – min)/(max – min)).

```python
# Create a 3×3 array with values 1 → 9
arr = np.arange(1, 10).reshape(3, 3)
print("Array:\n", arr)

# Compute row-wise and column-wise means
row_mean = arr.mean(axis=1)
col_mean = arr.mean(axis=0)

print("\nRow-wise means:", row_mean)
print("Column-wise means:", col_mean)

# Simulate 100 dice rolls (values between 1 and 6)
rolls = np.random.randint(1, 7, size=100)

# Count how many times each number appears
counts = np.bincount(rolls)[1:]  # ignore index 0
print("\nDice roll counts (1–6):", counts)

# Normalize an array so all values lie between 0 and 1
normalized_arr = (arr - arr.min()) / (arr.max() - arr.min())
print("\nNormalized array:\n", normalized_arr)
```

## Matplotlib & Visualization

### What is Matplotlib?

Matplotlib is Python’s standard plotting library.
It lets you visualize data to understand patterns and trends.

```python
import matplotlib.pyplot as plt
```

**Basic Line Plot**

```python
x = np.linspace(0, 10, 100)  # 100 points 0→10
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

**Multiple Lines and Legend**

```python
y_cos = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.title("Sine and Cosine")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

`plt.legend()` shows labels you define with `label=`.

**Scatter, Bar, and Histogram Plots**

```python
# Scatter plot
plt.scatter(x, np.sin(x))
plt.title("Scatter Plot of Sine")
plt.show()

# Bar chart
students = ["Alice","Bob","Charlie","Diana"]
scores = [85, 90, 78, 92]
plt.bar(students, scores)
plt.title("Student Scores")
plt.ylabel("Score")
plt.show()

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram of Random Data")
plt.show()
```

**Subplots and Saving Figures**

```python
x = np.linspace(0, 5, 50)
y1 = x**2
y2 = np.sqrt(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("x squared")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Square root of x")

plt.tight_layout()
plt.savefig("plots_example.png")  # saves the figure
plt.show()
```

`plt.subplot(rows, cols, index)` creates multiple plots in one figure.

`plt.savefig()` stores your figure as a file (e.g., PNG or PDF).

### In this exercise, we create a histogram of 1000 random numbers using Matplotlib to visualize their distribution. The bins parameter in the plt.hist() function controls how the data range is divided into intervals (bins). By changing the number of bins, we can see how the appearance of the histogram changes.

### If we use too few bins, the histogram looks too coarse and hides important details of the data distribution. On the other hand, using too many bins can make the histogram too noisy, making it hard to see the overall trend.

### To choose the best number of bins, we can experiment with different values (e.g., 10, 20, 30, 50, 100) and observe which gives a clear, balanced view of the data. There are also mathematical rules such as Sturges’ rule, Rice rule, or Freedman–Diaconis rule that help estimate a good number of bins automatically, but visual experimentation is often the simplest and most intuitive way.

```python
# Histogram of 1000 random numbers
data = np.random.randn(1000)

plt.figure(figsize=(6, 4))
plt.hist(data, bins=10, color="purple", edgecolor="white", alpha=0.8)
plt.title("Histogram of 1000 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
